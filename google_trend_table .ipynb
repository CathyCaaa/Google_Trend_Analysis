```python
from pytrends.request import TrendReq
from pprint import pprint
pytrend = TrendReq(hl='en-US', tz=360)
import pandas as pd
import numpy as np
import datetime,time 
import xlsxwriter
```

```python
data = pd.read_excel(r'排片大表TW     Taiwan Programming Schedule.xlsx', sheet_name = 'TW New Release')
#for loop check writer path
path = "C:/Users/cathychen/GoogleTrendCounter.txt"
Data = pd.DataFrame()
```

```python
data = pd.DataFrame(data, columns = ['TVODDiscontinuedDate','Chinese Title Name','HotPicks'])
data.head(5)
```

```python
###Data Preprocessing
HotPicks = data[data.HotPicks == "V"]
HotPicks = HotPicks.drop_duplicates('Chinese Title Name', keep = 'last')
today = datetime.date.today().strftime("%Y-%m-%d %H:%M:%S")
HotPicks['TVODDiscontinuedDate'] = HotPicks['TVODDiscontinuedDate'].apply(str)
HotPicks = HotPicks[HotPicks.TVODDiscontinuedDate > today]
HotPicks.info()
```

```python
keywords = pd.DataFrame(HotPicks, columns = ['Chinese Title Name'])
keywords = keywords.iloc[0:200,:]
```

```python
keywords = keywords.values.tolist()
keywords
```

```python
dicti = {}
i = 1
print("Start : %s" % time.ctime())
for trending in keywords:    
    #time.sleep(60)
    pytrend.build_payload(
    kw_list=trending,
    cat=34, # Movies , category definition:https://github.com/pat310/google-trends-api/wiki/Google-Trends-Categories
    timeframe='2020-01-01 2020-04-30',
    geo='TW',
    gprop='')
    dicti[i] = pytrend.interest_over_time()
    i+=1

#     result.to_csv('Google Trend.csv', mode = 'a',encoding = 'utf-8')
    
print("End : %s" % time.ctime())
```

```python
result = pd.concat(dicti, axis=1)
result.columns = result.columns.droplevel(0)
result = result.drop('isPartial', axis = 1)
result
```
